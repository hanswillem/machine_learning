# Find the tangent line for a point on a curve using the derivative

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

The formula we are starting with is:

$f(x) = x^2$

```python
def f(x):
    return x ** 2
```

```python
x = np.linspace(-4, 4, 100)
y = f(x)
```

```python
plt.style.use('ggplot')

plt.xlim(-4, 4)
plt.ylim(-4, 16)

plt.axhline(0, color='gray')
plt.axvline(0, color='gray')

plt.plot(x, y)
plt.show()
```

Now we calculate the derivative of $f(x)$:

$f'(x) = 2x$

Then we take the slope function:

$y - y_1 = m(x - x_1)$ 

Here $m$ is equal to $f'(x)$ for a point ($x_1$, $y_1$) on $f(x)$. So in our case this becomes:

$y - y_1 = 2x_1(x - x_1)$ 

We can rewrite it as follows:

$y = 2x_1(x - x_1) + y_1$

```python
def f_tangent(x_tangent, x_curve):
    y_curve = f(x_curve)
    return 2 * x_curve * (x - x_curve) + y_curve
```

Now we make an array of data for the tangent.

```python
xt = np.linspace(-4, 4, 100)
```

And we plug in an $x$ value. This is an $x$ value for a point on $f(x)$.

```python
yt = f_tangent(xt, 2)
```

And finally we plot the result.

```python
plt.style.use('ggplot')

plt.xlim(-4, 4)
plt.ylim(-4, 16)

plt.axhline(0, color='gray')
plt.axvline(0, color='gray')

plt.plot(x, y)
plt.plot(xt, yt)
plt.show()
```
