# imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

### generate a 1000 random samples within a normal distribution and plot them

```python
r = np.random.randn(1000)
plt.plot(r, '.')
plt.show()
```

### make a histogram

```python
plt.hist(r, 255)
plt.show()
```

### show the mean and the standard deviation

```python
r.mean()
```

```python
r.std()
```
